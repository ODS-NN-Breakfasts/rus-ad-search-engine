```python
from ruwordnet import RuWordNet
wn = RuWordNet()
```

```python
synsets = wn.get_synsets("–∫—É—Ä—Ç–∫–∞")
for s in synsets:
    print(s)
synset = synsets[0]
print()
print(synset.hypernyms)
print()
print(synset.hyponyms)
print()
print([synonym.name for synonym in synset.senses])
```
